```python
# GET CONTROL VALUES TO CSV

def getControl(INPUTS: list, WIRES: list, infile: str, outfile: str, num_samples: int):
    # control_df = pd.DataFrame(np.zeros((len(WIRES), len(INPUTS))), index=WIRES, columns=INPUTS)
    tt = open(infile)
    control_vals = []
    for i, inwire in enumerate(INPUTS):
        print("CONTROL VALUE FOR INPUT NUMBER " + str(i))
        control = [0 for i in WIRES]
        for index in range(0, num_samples):
            # with open(filename) as tt:

            temp = tt.read(len(INPUTS))
            line0 = tt.read(len(WIRES))
            temp = tt.read(1)
            temp = tt.read(len(INPUTS))
            line1 = tt.read(len(WIRES))
            # while 'x' in line0:
            #     loc = line0.find('x')
            #     line0 = line0[:loc] + '0' + line0[loc+1:]
            #     line1 = line1[:loc] + '0' + line1[loc+1:]
            # line0_val = int(line0, 2)
            # line1_val = int(line1, 2)
            line0_val, line1_val = removex(line0, line1)
            # xor = line0_val ^ line1_val
            xor = bin(line0_val ^ line1_val)[2:].zfill(len(WIRES))
            xor = [int(k) for k in xor]  # [0, 1, 0, 0, 0, 0]
            control = [val + xor[index] for index, val in enumerate(control)]
            # control = 0
            # while xor:
            #     control += 1
            #     xor &= xor - 1
            temp = tt.read(1)
        control_vals.append(control)

    control_df = pd.DataFrame(control_vals, index=INPUTS, columns=WIRES)
    control_df.to_csv(outfile)


def removex(line0: str, line1: str):
    while 'x' in line0:
        loc = line0.find('x')
        if line1[loc] == 'x':
            line0 = line0[:loc] + '0' + line0[loc+1:]
            line1 = line1[:loc] + '0' + line1[loc+1:]
        elif line1[loc] == '0':
            line0 = line0[:loc] + '1' + line0[loc+1:]
        elif line1[loc] == '1':
            line0 = line0[:loc] + '0' + line0[loc+1:]
    while 'x' in line1:
        loc = line1.find('x')
        if line0[loc] == '0':
            line1 = line1[:loc] + '1' + line1[loc+1:]
        elif line0[loc] == '1':
            line1 = line1[:loc] + '0' + line1[loc+1:]
    return int(line0, 2), int(line1, 2)
```

```python
# GET HEURISTICS FROM CONTROL TABLW
def heuristic(num_samples: int, WIRES:list, inCSV: str, outCSV: str):
    control_heuristic = []
    control_df = pd.read_csv(inCSV, index_col=0)
    control_df = control_df.div(num_samples)
    for col in control_df:
        
        control_heuristic.append([control_df[control_df[col] != 0][col].mean(), control_df[control_df[col] != 0][col].median()])
    heuristic_df = pd.DataFrame(control_heuristic, index=WIRES, columns=['mean', 'median'])
    heuristic_df.to_csv(outCSV)

def markSusWires(inCSV: str, threshold: float = 0.001):
    heuristic_df = pd.read_csv(inCSV, index_col=0)
    heuristic_df = heuristic_df.fillna(1).T
    sus_wires_mean = []
    sus_wires_median = []
    sus_wires_both = []
    print(heuristic_df.loc['mean', 'CRC_OUT_1_0'])
    for wire in heuristic_df:
        if heuristic_df.loc['mean', wire] < threshold:
            sus_wires_mean.append(wire)
        if heuristic_df.loc['median', wire] < threshold:
            sus_wires_median.append(wire)
        if heuristic_df.loc['mean', wire] < threshold and heuristic_df.loc['median', wire] < threshold:
            sus_wires_both.append(wire)
    # print(heuristic_df)
    

# markSusWires('test.csv')
heuristic(2048, s35932_net.WIRES, 'controltables/s35932_control.csv', 's35932_heuristic.csv')
```
